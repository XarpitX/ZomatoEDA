```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
## to chech data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
#finding number of countries and records related to them
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
##pie chart -- top 3 countries that use zomato based on transaction
plt.pie(country_values[:3],labels = country_names[:3], autopct="'%1.2f%%'" )
```

```python
##Obervation: Zomato max records or transaction are from India
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
## observation:
1. when rating is between 4.5 to 4.9 ---- Excellent
2. when rating is between 4.0 to 4.4 ---- very good
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)

sns.barplot(x = 'Aggregate rating', y ='Rating Count', data = ratings)
```

```python

sns.barplot(x = 'Aggregate rating', y ='Rating Count',hue='Rating color', data = ratings, palette=['white','red','orange','yellow','green','green'])
```

```python
 #count plot
sns.countplot(x='Rating color', data = ratings, palette=['white','red','orange','yellow','green','green'])
```

```python
## find the countries that have given 0 rating 
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

```python
final_df.groupby(['Aggregate rating', 'Country']).size().reset_index().head(5)
```

```python
# which currency is used by which country
final_df.groupby(['Currency', 'Country']).size().reset_index()
```

```python
#which country has online delivery
final_df[final_df['Has Online delivery']== 'Yes'].groupby('Country').size().reset_index()
```
